```python
# importing required libraries for webscraping
import os
import selenium
import numpy as np
import pandas as pd
import urllib.request
from time import sleep
from random import randint
from selenium import webdriver as wd
```

```python
# SCRAPING URLS OF SAREE images
csv = ('saree')
product={}
product['image_urls']=[]
page= np.arange(1,4,1)
driver=wd.Chrome('chromedriver.exe')
for i in page:
    driver.get('https://www.amazon.in/s?i=apparel&bbn=1968256031&rh=n%3A1571271031%2Cn%3A1571272031%2Cn%3A1953602031%2Cn%3A1968253031%2Cn%3A1968256031&s=popularity-rank&lo=image&page=2&pf_rd_i=1968253031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=a9066e0f-f26e-4568-9230-26d6a7e1af78&pf_rd_r=7P087R8206FZM9VG5JAG&pf_rd_s=merchandised-search-5&qid=1593698389&ref=sr_pg_'+str(i))
    images=driver.find_elements_by_xpath("//img[@class='s-image']")
    for image in images:
        source =image.get_attribute('src')
        product['image_urls'].append(source)
    sleep(randint(1,10))
df = pd.DataFrame(product)
df.to_csv(csv+'.csv')
driver.quit()
```

```python
#Feching images from urls saved in csv
def url_tO_jpg(i,url,file_path):
    
    filename = 'saree-{}.jpg'.format(i)
    fullpath = '{}{}'.format(file_path,filename)
    urllib.request.urlretrieve(url,fullpath)
    print('{} saved'.format(filename))
    
    return None

Filename = 'saree.csv'
target_folder = os.path.join('web/training_set/saree/')
if not os.path.exists(target_folder):
    os.makedirs(target_folder)
urls = pd.read_csv(Filename)

for i,url in enumerate(urls.values):
    if i<120:
        url_tO_jpg(i,url[1],target_folder)
```
